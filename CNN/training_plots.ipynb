```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
```

```python
epoch,train,test=np.loadtxt('weightemit_ae0_vel_loss.txt',skiprows=1,delimiter=',',unpack=True)
fig, ax = plt.subplots(figsize=(6,4),dpi=400)
plt.plot(epoch,train,color='k',label='train')
plt.plot(epoch,test,color='k',linestyle=':',label='test')
plt.xlim(0,100)
plt.ylim(0,610)
plt.vlines(10.5,0,700,color='r',label='unfreeze encoder weights')

epoch_best = np.argmin(test) + 1
plt.scatter(epoch_best,np.min(test),color='blue',zorder=10,s=80,label='min test loss')
plt.xlabel('epoch')
plt.ylabel('loss')

plt.legend(loc='lower right')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4),dpi=200)
ax.plot(epoch,train,color='k',label='train')
ax.plot(epoch,test,color='k',linestyle=':',label='test')
ax.set_xlim(1,100)
ax.set_ylim(150,650)
ax.vlines(10.5,0,700,color='r',label='unfreeze encoder weights')

epoch_best = np.argmin(test) + 1
ax.scatter(epoch_best,np.min(test),color='blue',zorder=10,s=80,label='min test loss')
ax.set_xlabel('epoch')
ax.set_ylabel('loss')

# --- animation update ---
N_FRAMES = 150  # frames in the GIF
xlim_set = np.append(np.linspace(1.5,20,N_FRAMES),np.linspace(20,100,N_FRAMES))

def update(frame):
    # frame goes from 0 â†’ N_FRAMES-1
    ax.set_xlim(1, xlim_set[frame])
    return ax,

ani = FuncAnimation(fig, update, frames=2*N_FRAMES, interval=50, blit=False)

# --- save as GIF ---
writer = PillowWriter(fps=20)
ani.save("loss_expand.gif", writer=writer)

fig.show()
```
